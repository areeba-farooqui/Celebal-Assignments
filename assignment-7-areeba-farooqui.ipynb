```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
documents = [
    "Artificial Intelligence is transforming many industries.",
    "Machine Learning is a subset of Artificial Intelligence.",
    "Deep Learning uses neural networks with many layers.",
    "RAG stands for Retrieval Augmented Generation.",
    "Vector databases store embeddings for similarity search."
]
```

```python
vectorizer = TfidfVectorizer()

doc_vectors = vectorizer.fit_transform(documents)

print(doc_vectors.shape)
```

```python
query = "What is RAG?"

query_vector = vectorizer.transform([query])
```

```python
similarities = cosine_similarity(query_vector, doc_vectors)

top_indices = np.argsort(similarities[0])[::-1][:2]

print(top_indices)
```

```python
print("Question:", query)

print("\nRetrieved Documents:")

for idx in top_indices:
    print("-", documents[idx])
```

```python
def ask_question(query):
    
    query_vector = vectorizer.transform([query])

    similarities = cosine_similarity(query_vector, doc_vectors)

    top_index = np.argmax(similarities)

    print("Question:", query)
    print("\nAnswer:")
    print(documents[top_index])
```

```python
ask_question("What is RAG?")
```

```python
ask_question("What is Machine Learning?")
```

# Conclusion

This project demonstrates a simple Retrieval-Augmented Generation (RAG) pipeline.

Steps implemented:
1. Document ingestion
2. Text vectorization using TF-IDF

5. Generation of answers from retrieved context
